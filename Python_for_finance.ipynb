```python
import numpy as np
import numpy_financial as npf

import matplotlib.pyplot as plt 

#price for 2018-2021
bitcoin = [3869.47, 7188.46, 22203.31, 29391.78]
# initial_investment_value = 500000

def calculate_array_for_irr(bitcoin, value, multiplier): 
    bitcoin = np.multiply(bitcoin, multiplier)
    return npf.irr(np.insert(bitcoin, 0, -value))

#print(calculate_array_for_irr(bitcoin,initial_investment_value, 10))



def calculate_investment_worth(bitcoin, initial_investment):
    return np.multiply(bitcoin, initial_investment)



def investment_worth_plot():

    initial_investment = 1000
    investment_worth = calculate_investment_worth(bitcoin, initial_investment)
    years = [2018,2019,2020, 2021]


    plt.plot(years, investment_worth)
    plt.savefig('investment_worth.png')



import yfinance as yf 



def bitcoin_price():
    data = yf.Ticker('BTC-USD')
    x = data.history('1y')['Close']
    start = 1000 / x[0]
    y = np.multiply(start, x)
    plt.plot(y)
    plt.savefig('plot.png')

#bitcoin_price()


def bitcoin_volatility(bitcoin):
    bitcoin_volatility = np.std(bitcoin)
    print("Risk:",bitcoin_volatility)


def bitcoin_sharpe(bitcoin):
    annual_std = np.std(bitcoin) * np.sqrt(252)
    sharpe = (np.mean(bitcoin) / np.std(bitcoin))*np.sqrt(252)
    print("Sharpe:", sharpe)

bitcoin_volatility(bitcoin)
bitcoin_sharpe(bitcoin)

data = yf.download("BTC-USD", start='2021-01-01') 
data['Close'].plot()
plt.savefig('plot.png')

```
